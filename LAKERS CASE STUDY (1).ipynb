```python
def read_multiple_csv(path):
    csv_files=[f for f in os.listdir(path) if f.endswith('.csv')]
    dataframes={}
    
    for csv_file in csv_files:
        dataframes[csv_file]=pd.read_csv(os.path.join(path,csv_file))
        
    return dataframes    
```
